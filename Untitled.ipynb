```python
ds = load_dataset("obalcells/longfact-augmented-prompts",streaming=False)

ds = ds.map(lambda x: x, batched=False)

def tokenize(example):
    messages = [
    {"role": "user", "content": f"{example['question']} Provide as many specific details and examples as possible (such as names of people, numbers, events, locations, dates, times, etc.)."},
    ]
    inputs = tokenizer.apply_chat_template(
    	messages,
    	add_generation_prompt=True,
    	tokenize=True,
    	return_dict=True,
    	return_tensors="pt",
      enable_thinking=False
    )
    example['input_ids'] = inputs['input_ids'][0].cpu().tolist()
    example['attention_mask'] = inputs['attention_mask'][0].cpu().tolist()

    return example

ds = ds.map(tokenize, batched=False)

    
ds.save_to_disk("./long_factpp")
```
